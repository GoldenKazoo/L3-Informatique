```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler

# Standard
scaler_std = StandardScaler()
X_train_std = scaler_std.fit_transform(X_train)
X_test_std = scaler_std.transform(X_test)

# MinMax
scaler_mm = MinMaxScaler()
X_train_mm = scaler_mm.fit_transform(X_train)
X_test_mm = scaler_mm.transform(X_test)
```


**Appliquer  la méthode du k plus proche voisins**
k=2,3,4

1- sans faire le prétraitement 

2- appliquer d'abord StandardScaler puis KNeighborsClassifier(n_neighbors =k)

3- appliquer d'abord MaxMinScalar puis KNeighborsClassifier(n_neighbors =k)

Comparez les résultats

```python
from sklearn.neighbors import KNeighborsClassifier

for k in [2,3,4]:
    print("k =", k)

    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    print("Sans scaling:", knn.score(X_test, y_test))

    knn_std = KNeighborsClassifier(n_neighbors=k)
    knn_std.fit(X_train_std, y_train)
    print("Standard:", knn_std.score(X_test_std, y_test))

    knn_mm = KNeighborsClassifier(n_neighbors=k)
    knn_mm.fit(X_train_mm, y_train)
    print("MinMax:", knn_mm.score(X_test_mm, y_test))
    print()
```
